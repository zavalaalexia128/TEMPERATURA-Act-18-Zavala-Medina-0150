### 1. Fórmulas de ConversiónPara movernos entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:De Celsius ($^\circ C$) a Fahrenheit ($^\circ F$):$$^\circ F = (^\circ C \times \frac{9}{5}) + 32$$(O de forma decimal: $^\circ F = (^\circ C \times 1.8) + 32$)De Fahrenheit ($^\circ F$) a Celsius ($^\circ C$):$$^\circ C = \frac{^\circ F - 32}{1.8}$$
### 2. Ejemplos Resueltos Paso a PasoEjemplo A: Convertir $25^\circ C$ a FahrenheitObjetivo: Saber a cuánto equivalen $25^\circ C$ (una temperatura ambiente agradable).Sustitución: Usamos la fórmula $F = (C \times 1.8) + 32$.$$F = (25 \times 1.8) + 32$$Multiplicación: Multiplicamos $25$ por $1.8$.$$25 \times 1.8 = 45$$Suma: Sumamos $32$ al resultado anterior.$$45 + 32 = 77$$Resultado: $25^\circ C$ equivalen a $77^\circ F$.Ejemplo B: Convertir $98.6^\circ F$ a CelsiusObjetivo: Convertir la temperatura corporal promedio de Fahrenheit a Celsius.Sustitución: Usamos la fórmula $C = (F - 32) / 1.8$.$$C = (98.6 - 32) / 1.8$$Resta: Primero resolvemos lo que está dentro del paréntesis.$$98.6 - 32 = 66.6$$División: Dividimos el resultado entre $1.8$.$$66.6 / 1.8 = 37$$Resultado: $98.6^\circ F$ equivalen a $37^\circ C$.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    try:
        celsius = float(input("\nIngrese los grados Celsius (°C): "))
        fahrenheit = (celsius * 1.8) + 32
        print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    try:
        fahrenheit = float(input("\nIngrese los grados Fahrenheit (°F): "))
        celsius = (fahrenheit - 32) / 1.8
        print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def menu():
    """Controla el flujo principal del programa"""
    while True:
        print("\n=== MENÚ DE CONVERSIONES DE TEMPERATURA ===")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Zavala Medina NL = 54")
            break
        else:
            print("Opción no válida. Intente de nuevo.")
if __name__ == "__main__":
    menu()
```
